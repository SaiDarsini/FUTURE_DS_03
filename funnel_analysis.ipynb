```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:/Future_DS_03/data/raw/Ecommerce_simpleData.csv")

print(df.shape)
df.head()
```

```python
df.columns
```

```python
funnel_counts = df['event_type'].value_counts()
print(funnel_counts)
```

```python
views = funnel_counts.get('view', 0)
carts = funnel_counts.get('cart', 0)
purchases = funnel_counts.get('purchase', 0)

view_to_cart = carts / views
cart_to_purchase = purchases / carts
overall_conversion = purchases / views

print("View → Cart:", round(view_to_cart*100,2), "%")
print("Cart → Purchase:", round(cart_to_purchase*100,2), "%")
print("Overall Conversion:", round(overall_conversion*100,2), "%")
```

```python
drop_view_cart = 1 - view_to_cart
drop_cart_purchase = 1 - cart_to_purchase

print("Drop-off at Cart Stage:", round(drop_view_cart*100,2), "%")
print("Drop-off at Purchase Stage:", round(drop_cart_purchase*100,2), "%")
```

```python
channel_conversion = df[df['event_type']=="purchase"]['traffic_source'].value_counts()
print(channel_conversion)
```

```python
funnel_summary = pd.DataFrame({
    "Stage": ["View","Cart","Purchase"],
    "Count": [views, carts, purchases]
})

funnel_summary.to_csv("D:/Future_DS_03/data/processed/funnel_summary.csv", index=False)
```
